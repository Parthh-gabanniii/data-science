```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn import set_config
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Fare','Sex','Survived']]
df
```

```python
x = df.iloc[:,0:6]
x
```

```python
y = df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.5,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python

tnf1 = ColumnTransformer([('impute_farr',SimpleImputer(),[4])
                         ],remainder = 'passthrough')
tnf1
```

```python
tnf2 = ColumnTransformer([
  ('impute_sex',OneHotEncoder(sparse=False,handle_unknown='ignore'),[5])  
],remainder = 'passthrough')
tnf2
```

```python
pipe = make_pipeline(tnf1,tnf2)
```

```python
pipe.fit(x_train,y_train)
```

```python

set_config(display='diagram')
```

```python
y_pred = pipe.transform(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
