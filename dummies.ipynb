```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('LORdataset.csv')
df
```

```python
new = pd.get_dummies(df['Review'])
new
```

```python
drop_column = df.drop('Review',axis = 'columns')
drop_column
```

```python
new_df = pd.concat([new,drop_column],axis='columns')
new_df
```

```python
x = new_df.drop('Average_Marks',axis = 'columns')
x
```

```python
y = new_df.Average_Marks
y
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr.fit(x,y)
```

```python
lr.predict([[0,1]])
```
